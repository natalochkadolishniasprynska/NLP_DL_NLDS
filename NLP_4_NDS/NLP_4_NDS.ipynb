```python
import nltk
from nltk.corpus import gutenberg
import numpy as np

def create_co_occurrence_matrix(docs, window_size=2):
    vocab = set()
    word2index = {}
    for doc in docs:
        for word in doc:
            if word not in vocab:
                word2index[word] = len(vocab)
                vocab.add(word)
    co_occurrence_matrix = np.zeros((len(vocab), len(vocab)))

    for doc in docs:
        for i, word in enumerate(doc):
            context = doc[max(0, i - window_size):i] + doc[i + 1:i + 1 + window_size]
            for context_word in context:
                co_occurrence_matrix[word2index[word]][word2index[context_word]] += 1

    return co_occurrence_matrix

def calculate_PPMI(co_occurrence_matrix):
    PPMI_matrix = np.zeros(co_occurrence_matrix.shape)
    row_totals = co_occurrence_matrix.sum(axis=1)
    col_totals = co_occurrence_matrix.sum(axis=0)
    total_sum = co_occurrence_matrix.sum()

    for i in range(co_occurrence_matrix.shape[0]):
        for j in range(co_occurrence_matrix.shape[1]):
            pij = co_occurrence_matrix[i][j] / total_sum
            pi_star = row_totals[i] / total_sum
            p_star_j = col_totals[j] / total_sum

            PPMI_matrix[i][j] = max(np.log2(pij / (pi_star * p_star_j)), 0)

    return PPMI_matrix

docs = ["melville-moby_dick.txt", "austen-sense.txt", "bryant-stories.txt"]
co_occurrence_matrix = create_co_occurrence_matrix(docs, window_size=2)
PPMI_matrix = calculate_PPMI(co_occurrence_matrix)

print(PPMI_matrix)
```

```python
epsilon = 1e-8

def calculate_smoothed_PPMI(co_occurrence_matrix):
    PPMI_matrix = np.zeros(co_occurrence_matrix.shape)
    row_totals = co_occurrence_matrix.sum(axis=1)
    col_totals = co_occurrence_matrix.sum(axis=0)
    total_sum = co_occurrence_matrix.sum()

    for i in range(co_occurrence_matrix.shape[0]):
        for j in range(co_occurrence_matrix.shape[1]):
            pij = co_occurrence_matrix[i][j] / total_sum
            pi_star = (row_totals[i] + epsilon) / (total_sum + epsilon)
            p_star_j = (col_totals[j] + epsilon) / (total_sum + epsilon)

            PPMI_matrix[i][j] = max(np.log2(pij / (pi_star * p_star_j)), 0)

    return PPMI_matrix

docs = ["melville-moby_dick.txt", "austen-sense.txt", "bryant-stories.txt"]
co_occurrence_matrix = create_co_occurrence_matrix(docs, window_size=5)  # Choose window size here
PPMI_matrix = calculate_smoothed_PPMI(co_occurrence_matrix)

print(PPMI_matrix)
```
